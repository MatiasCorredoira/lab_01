## Análisis exploratorio

### Correlaciones entre variables

```python
LaLiga.corr()
```

#### Principalmente llamativas son 2 correlaciones:
#### - Entre goles marcados y el número de camiseta (correlación negativa) esto es debido a que la mayoria de futbolistas que juegan más minutos llevan dorsales entre el 1 y el 11 que son el número de jugadores en terreno de juego.
#### - Entre el nº de corners y los duelos hombre-hombre(particularmente alta) esto es debido a que los equipos que en mayor medida buscan tener corners suelen confeccionar sus plantillas de modo que sus jugadores sean particularmente fuertes, lo cual posee importancia cada vez mayor en el fútbol de élite.

### Búsqueda del jugador con dorsal mayor

```python
nº_dorsal= (LaLiga.groupby('Shirt number').agg(suma_goles_cabeza = ('Shirt number','count',)))
nº_dorsal
```

#### Llama mucho la atención la caída en el número de jugadores a partir del número 25. Esto es debido a que la Liga De Fútbol Profesional solo permite llevar estos números a jugadores que proceden del 2º equipo, los cuales evidentemente son muy pocos en relación al resto, el dorsal mayor es el número 57.

### Búsqueda del jugador con dorsal más alto

```python
dorsales = LaLiga['Shirt number']==57
dorsales_nombre=LaLiga[dorsales]
dorsales_nombre.Name
```

### Búsqueda de jugadores que disputan más de 30 partidos completos

```python
LaLiga_jugadores1=LaLiga[LaLiga["Full games played"]>=30]
LaLiga_jugadores1.sort_values("Full games played",ascending=False)
```

#### Las primeras posiciones están ocupadas por porteros debido a que en esa posición se da una rotación de jugadores mucho menor en relación al resto. Un escalón por debajo se encuentran los defensas (que en su mayoría pertenecen a equipos de la zona media de la tabla) Esto es debido a que esos equipos suelen cuidar más la parcela defensiva y por ello rotan menos a sus jugadores más importantes.

### Jugadores que disputan menos de 2 partidos

```python
LaLiga_partidos_1=LaLiga[LaLiga["Games played"]<2]
LaLiga_partidos_1
```

#### En la tabla podemos ver de nuevo la presencia de jugadores con dorsales mayores a 25, esto es debido a lo que comentamos anteriormente. Los jugadores que pertenecen al 2º en muchas ocasiones juegan de modo circunstancial por falta de jugadores en el primer equipo, o bien por probarlos y ver su nivel. En la gran mayoría de casos estos no consiguen afianzarse, prueba de la exigencia de la liga en España.

### Diagramas de caja del número de camiseta

```python
sns.boxplot(x=LaLiga["Position"],y=LaLiga["Shirt number"])
```

#### Llaman la atención 2 hechos:
#### - La menor varianza por parte de los porteros y los delanteros, esto es debido a la menor cantidad de jugadores que cubren estas posiciones, aunque también es digno de mención que estas posiciones hay una fijación particular por ciertos números como es el caso del 1 para los porteros o el 9 para los delanteros.
#### - Los valores extremos, podemos comprobar que la posición en la que han jugado más jugadores provenientes de la segunda plantilla son mediocampistas, esto es debido en gran medida a 2 factores: 
####    1º) El mayor desgaste de los jugadores, ya que el centro del campo es la parcela en la que las disputas de balón son más frecuentes.
####    2º) El fútbol de élte cada vez exige a sus futbolistas un conocimiento general mayor de todos los aspectos del juego. Para ello los jugadores que juegan en una zona intermedia por condición tienen menos complicado mecanixar todos los automatismos del juego.

### Equipos que marcan más goles de cabeza

```python
goles_cabeza = (LaLiga.groupby('Team').agg(suma_goles_cabeza = ('Goals scored with header','count')))
goles_cabeza
```

### Equipos que participan en más duelos

```python
jugadores_mas_duelos = (LaLiga.groupby('Team').agg(suma_duelos_parejos =('Man-to-man duels','sum')))
jugadores_mas_duelos
```

#### De nuevo podemos ver otra correlación semejante a una de las que se habló anteriormente. Los equipos que marcan más goles de cabeza de modo general son equipos que buscan mucho duelo 1 para 1. Esto es debido a que rematar de cabeza sin que haya disputa resulta más complicado, ya que la recepción de balones por alto es más complicada que la de balones rasos.

```python
posiciones_mas_duelos= (LaLiga.groupby('Position').agg(suma_de_duelos_disputados =('Man-to-man duels','sum')))
posiciones_mas_duelos
```

#### Existe una clara diferencia entre los duelos entre centrocampistas en relación al resto, debido al motivo que se ha explicado anteriormente.

```python
plt.scatter(LaLiga["Long passes"],LaLiga["Team"])
plt.ylabel("Long passes")
plt.xlabel("Team")
plt.show()
```

#### De nuevo se plasma lo comentado antes, los equipos que buscan más pases en largo (caso de  Atheltic Club o Getafe) Son equipos que suelen jugar de un modo de juego más físico.

### Aquí lo vemos representado en un gráfico

```python
fig, ax = plt.subplots(nrows = 1, figsize=(25, 5)) 

ax = sns.lineplot(x='Team',y='Long passes',data=LaLiga, label='Long passes')

ax.set(ylabel = 'Cifras')
ax.legend()
plt.show()
```

#### Existen valores más altos en los equipos que juegan con más frecuencias contransiciones rápidas en ataque como es el caso del Eibar o del Getafe, o también el Atlético de Madrid ya mencionado en anteriores ocasiones. 

### Grafica de tarjetas amarillas

```python
fig, ax = plt.subplots(nrows = 1, figsize=(25, 5)) 

ax = sns.lineplot(x='Team',y='Yellow Cards',data=LaLiga, label='Yellow Cards')

ax.set(ylabel = 'Equipos')
ax.legend()
plt.show()
```

#### Juntando esta gráfica con la anterior se puede ver la existencia de correlación positiva entre las tarjetas amarillas y los pases largos. Esto es debido a que los equipos que realizan más pases largos tienden a dedicarse a defender(generalmente en bloque bajo) mucho más tiempo que a atacar en sus partidos. Por lo que es más probable que realicen más faltas ya que disponen de menos tiempo de posesión de balón.

 ### Grafica goles marcados-tarjetas amarillas-tarjetas rojas

```python
fig, ax = plt.subplots(nrows = 1, figsize=(25, 5)) 

ax = sns.lineplot(x='Team',y='Goals scored',data=LaLiga, label='Goals scored')
ax = sns.lineplot(x='Team',y='Yellow Cards',data=LaLiga, label='Yellow Cards')
ax = sns.lineplot(x='Team',y='Red Cards',data=LaLiga, label='Red Cards')
ax.set(ylabel = '')
ax.legend()
plt.show()
```

#### En este gráfico se observa una diferencia clara respecto a los goles y las tarjetas amarillas. Los equipos que suelen ocupar las posiciones más altas de la tabla como el Madrid, Barcelona, Valencia o  Atlético reciben menos tarjetas amarillas debido a que suelen llevar la iniciativa del juego. Por lo que tienen más posesión de balón durante la cual no realizan faltas. Esto en el resto de equipos no lo podemos apreciar de la misma manera.
#### Respecto a las tarjetas rojas no se puede apreciar del mismo modo la diferencia debido a que estas se dan en menor cantidad, sin embargo el hecho que comentamos respecto a las tarjetas amarillas se da en mayor en relación a las tarjetas rojas. El principal objetivo de esta gráfica además de mostrar los datos mencionados es mostrar un ejemplo de como se puede engañar mediante la visualización datos verídicos.

### Displot De goles anotados

```python
sns.displot(LaLiga, x="Goals scored")
```

#### De forma clara se observa una correlación negativa. La mayoría de jugadores de hecho ni siquiera llegan a notar por lo que se percibe una gran concentración de los goles. Es digno de mención la diferencia abismal que se encuentra entre el primer jugador (Leo Messi) y la segunda posición, compartida por Luis Suárez y Karim Benzemá, ambos con 21 goles.

### Dispersión goles-partidos comenzados

```python
sns.regplot(x='Games started',y='Goals scored',data=LaLiga,order=5)
plt.show()
```

#### Evidentemente al comenzar más partidos resulta más facil marcar más goles, sin embargo vemos que la linea de regresión tiene un ángula bastante bajo. Podemos contemplar en el valor más apartado los 36 goles de Leo messi disputando 29 partidos, o los 38 partidos de Diego López con 0 goles en contraposición.

### Gráfico de densidad de tarjetas amarillas

```python
plt.figure(figsize=(5, 6))
sns.distplot(LaLiga['Yellow Cards'], color='yellow', bins=50, hist_kws={'alpha': 0.005});
```

#### La gran mayoría de jugadores reciben claramente entre 0 y 5 tarjetas amarillas, el 100% de la distribución de jugadores se da en el punto 15 del eje x el cual supone el máximo número de tarjetas amarillas recibidas por un jugador.

### Gráfico de densidad de tarjetas rojas

```python
plt.figure(figsize=(5, 6))
sns.distplot(LaLiga['Red Cards'], color='red', bins=50, hist_kws={'alpha': 0.05});
```

#### Podemos observar la misma tendencia que en la anterior gráfica pero con valores mucho más concentrados. La gran mayoría de jugadores no llegan a recibir 1 mientras que el número máximo de las mismas, es solo 2.
